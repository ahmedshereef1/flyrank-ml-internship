## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

Research Question

Which content pages should be reviewed first for content refresh based on their search performance and engagement signals?

Decision

Determine which pages deserve immediate review.

Action

A content manager reviews the highest-ranked pages and decides whether to:

Update the content
Improve SEO
Expand the article
Monitor the page
Leave it unchanged
Unit of Analysis

One content page.

Output

A ranked list of pages with:

Refresh score
Confidence score
Reason codes explaining the recommendation
Cost of a Wrong Recommendation

If a page is incorrectly recommended for review, time and resources are wasted.

If an important page is missed, the business may lose valuable search traffic and potential customers.

Why Machine Learning?

Simple rules cannot capture the complex relationships between search visibility, user engagement, freshness, and content quality. Machine learning can combine multiple signals to produce better rankings than fixed rule-based systems while still providing explainable recommendations.
